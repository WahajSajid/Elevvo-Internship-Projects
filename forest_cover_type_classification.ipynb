```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import accuracy_score
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.metrics import SparseCategoricalCrossentropy
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.linear_model import SGDClassifier
```

```python
#load the dataset
data = pd.read_csv(rf"C:\Users\Wahaj Sajid\Downloads\archive (1)\covertype.csv")
data.head()
```

# Preprocess the data

```python
#checking for the null values in the data set
print(data.isnull().sum())
```

```python
#checking for the unique values in the "Cover_Type"

print(data['Cover_Type'].unique())
```

```python
# defining x and y
x = data.drop(columns=['Cover_Type'], axis=1)
y = data['Cover_Type']
```

```python
#train, test, split the model
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#scale the x
x_scaler = MinMaxScaler()
x_train = x_scaler.fit_transform(x_train)
x_test = x_scaler.transform(x_test)
```

```python
x_train = np.array(x_train)
x_test = np.array(x_test)

y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
classifier = SGDClassifier(loss= 'log_loss', max_iter=1000, tol= 0.05, random_state=42)
classifier.fit(x_train, y_train)
```

```python
#making predictions
train_predictions = classifier.predict(x_train)
test_preidcitons = classifier.predict(x_test)
print(f'Accuracy on Test data: ',accuracy_score(y_test, test_preidcitons))
print(f'Accuracy on Train data: ',accuracy_score(y_train, train_predictions))
```
